### 1.     

```python
import numpy as np
import pandas as pd

from tensorflow.keras.datasets.mnist import load_data

(train_x, train_y), (test_x, test_y) = load_data()
```

```python
print(train_x.shape, train_y.shape)

print(test_x.shape, test_y.shape)
```

### 2.  

```python
from PIL import Image
img = train_x[0]

import matplotlib.pyplot as plt
# fromarray : numpy  PIL   
# ,  Numpy img  8    (mode='L')
img1 = Image.fromarray(img, mode='L')
plt.imshow(img1)
train_y[0] # 0    
```

### 3.  

```python
#    
train_x1 = train_x.reshape(60000, -1)
test_x1 = test_x.reshape(10000, -1)
```

```python
train_x1[0]
```

```python
#   255 .
train_x2 = train_x1 / 255
test_x2 = test_x1 / 255
```

```python
train_x2[0]
```

### 4.  

```python
#     
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
#  md  (Dense) 
md = Sequential()
md.add(Dense(128, activation='relu', input_shape=(28*28,)))
md.add(Dense(64, activation='relu'))
md.add(Dense(10, activation='softmax')) #    softmax 
md.summary() #  
```

> softmax           ,      1  . (=      )

### 5.   

```python
#  compile
md.compile(loss='sparse_categorical_crossentropy', optimizer='sgd', metrics='acc')
```

> compile() :     .
> loss='sparse_categorical_crossentropy' :         . 

```python
#   : , batch_size, validation_split
hist = md.fit(train_x2, train_y, epochs=50, batch_size=64, validation_split=0.2)
```

> batch_size=64 :      64  . , 64   .
> validation_split=0.2 :  20%   .  20%       . 

```python
hist.history
```

```python
#    
acc = hist.history['acc']
val_acc = hist.history['val_acc']
epoch = np.arange(1, len(acc)+1)

plt.figure(figsize=(10,8))
# plt.xlim(250, len(acc)+1)
# plt.ylim(0.92, 0.94)

plt.plot(epoch, acc, 'b', label='acc')
plt.plot(epoch, val_acc, 'g', label='val_acc')
plt.legend()

plt.show()
```

> 

### 6. 

```python
md.evaluate(test_x2, test_y)
```

### 1) layer 

```python
#  
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

md = Sequential()
md.add(Dense(128, activation='relu', input_shape=(28*28, )))
md.add(Dense(64, activation='relu', input_shape=(28*28, )))
md.add(Dense(32, activation='relu', input_shape=(28*28, )))
md.add(Dense(10, activation='softmax'))

md.summary()
```

```python
#   
md.compile(loss='sparse_categorical_crossentropy', optimizer='sgd', metrics='acc')
hist = md.fit(train_x2, train_y, epochs=50, batch_size=64, validation_split=0.2)
```

```python
hist.history
```

```python
acc = hist.history['acc']
val_acc = hist.history['val_acc']
epochs = np.arange(1,len(acc)+1)

plt.figure(figsize=(10,8))
plt.plot(epochs, acc, 'b', label='acc')
plt.plot(epochs, val_acc, 'g', label='val_acc')
plt.legend()

plt.show()
```

>  . 

```python
md.evaluate(test_x2, test_y)
```

#### 2) dropout  

```python
#  
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

md = Sequential()
md.add(Dense(128, activation='relu', input_shape=(28*28, )))
md.add(Dense(64, activation='relu'))
md.add(Dropout(0.5))
md.add(Dense(10, activation='softmax'))
md.summary()
```

> Dropout         .  ,       .       ,      ,    . <br>
Dropout   (fully connected)   (convolutional)   .       ,     (batch) .         ,       . <br>
Dropout    ,     .       ,     . <Br>
 Dropout            .

```python
#  
md.compile(loss='sparse_categorical_crossentropy', optimizer='sgd', metrics='acc')
hist = md.fit(train_x2, train_y, epochs=50, validation_split=0.2, batch_size=64)
```

```python
hist.history
```

```python
acc = hist.history['acc']
val_acc = hist.history['val_acc']
epochs = np.arange(1, len(acc)+1)

plt.figure(figsize=(10,8))
plt.plot(epochs, acc, 'b', label='acc')
plt.plot(epochs, val_acc, 'g', label='val_acc')
plt.legend()

plt.show()
```

>   .

```python
#  
md.evaluate(test_x2, test_y)
```

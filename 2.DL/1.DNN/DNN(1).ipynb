## RNN 신경망 모델
= 선형모델 + 활성화함수(sigmoid, tanh, ReLu)

__신경망 모델과정__
1. 층이 하나인 선형모델에서 보다 더 층이 깊어진 계기가 신경망 모델.
2. 신경망의 층을 구분해주는 것은 비선형함수인 `활성화함수`
3. 층이 깊어지면서 최적화를 위한 미분 계산도 복잡해지는데 이를 해결한 것이 `오류역전파`
4. 학습의 한계를 나타내는 지표가 되는 것이 `과적합`
5. 과적합 해소방안인 `규제강화(Regularization) 및 SGD를 개선한 RMSProp, adam`등의 최적화 기법.




__층(Layer)의 결합__ <br>

1. 선형모델 <br>
선형모델은 입력 벡터에 행렬을 곱함으로써 모델이 만들어진다. 이런 선형 모델을 여러 개 결합하면 행렬과 행렬을 곱하는 것이므로, 여러 개를 결합해도 다시 선형모델이 될 뿐이다. (= 선형모델은 복잡한 데이터 패턴이나 관계를 모델링하는데 한계가 있따. 따라서 비선형모델로 변환해주어야 다양한 형태의 데이터에서 복잡한 패턴과 관계를 학습할 수 있게 된다.)

2. 선형모델 + 활성화 함수 = 비선형모델로! <br>

 <br>


신경망 모델은 `활성화 함수로 인해 층이 구분`된다. 행렬과 행렬 사이에 활성화 함수를 추가하는 방식으로 `여러 개의 행렬을 연결할 수 있다`. 다만 층이 많아질수록 모델이 복잡해지고 매개변수(parameter)의 개수가 많아지므로 계산도 복잡해진다. 


__활성화함수__ <br>



1. 활성화함수는 `선형모델에서 계산된 결과값을 활성화`시켜주는 역할을 한다. 
    - 주로 비선형적인 함수를 사용해 출력을 조절해주는 것으로, 이 과정을 통해 출력범위를 조절해 다음 층의 입력값의 범위를 설정할 수 있다. 이는 일종의 정규화 과정으로 볼 수 있다. 
    - 또한 단순한 선형 모델의 반복으로 훨씬 복잡한 표현이 가능하며, 층을 더 깊게 해 보다 정교한 최적화가 가능하도록 해준다.

    __활성화?__ <br>
    - 이전 층의 결과값이 다음 층으로 넘어가기 전에 값의 크기를 조절해주는 역할을 한다. 
    - 종류 : sigmoid, tanh, ReLu, swish..
        - sigmoid : 0~1사이 값으로 전처리할 때 주로 사용하는 값의 범위가 된다.
        - tanh : -1~1사이의 값으로 전처리할 때 주로 사용되는 값의 범위가 된다.
        - ReLu : max(x,0) 로 x가 양수이면, x그대로 출력하고 음수라면 0으로 출력하기에 출력값이 항상 0이상이 된다. 

2. 활성화 과정은 비선형함수로 이루어지기에 `미분이 가능`하며 학습이 가능하다. 층을 여러 개 쌓아 학습을 시키더라도 활성화 함수의 미분 가능성이 전체 학습의 흐름을 잘 이어지게 한다. 다만 활성화 함수의 함숫값이나 미분값이 학습에 영향을 미칠 수도 있기에 활성화 함수의 선택이 중요하다. `오류 역전파(error backpropagation)`란, 이처럼 여러층으로 된 모델의 최적화에 사용되는 기법이다. 


__Sigmoid 함수__





이진분류 문제의 경우 손실함수를 적용하기 전에 반드시 sigmoid활성화를 거친다. 이진분류는 결국 0이냐 1이냐의 문제이기 때문에, sigmoid로 활성화하면 출력값이 0과 1사이의 확률이 되기에 이진분류를 하기에 적합한 상태가 된다. sigmoid함수의 특성으로 인해 대부분의 값은 0또는 1에 가까운 값이 되며, 중간값은 보통 0.5이상인 경우는 1로, 미만인 경우는 0으로 처리한다.

Relu함수는 계산이 간단하고 학습도 잘되는 성능이 아주 우수한 활성화 함수이다. 단점으로는 입력값이 음수인 경우 학습이 전혀 이루어지지 않기 때문에 이를 보완한 활성화 함수들이 많이 고안되었따. 예를 들어 음수 영역에 기울기를 조금 주는 Leaky ReLu, 음수영역을 지수함수로 보완한 ELU, 그 외에도 Selu, mish 등의 함수들이 있다. 이 함수들은 양수 부분은 ReLu와 거의 같은 형태이지만, 음수 부분에 약간의 값을 주어 출력값이 0이 아니며, 미분값도 0이 아니기에 음수 구간에서도 조금이나마 학습이 가능하도록 설계된 활성화 함수들이다.

__학습곡선 LearinigCurve__ <br>
학습곡선은 학습이 진행되는 과정에서 정확도, 손실 등의 값을 저장해 그래프로 그린 것. 학습 과정에서 정확도, 손실값은 history에 저장된다. 

1) 학습 곡선에서 초반에 너무 변화가 없는 경우 : 모델 가중치 초기값 설정을 다시해야한다. <br>


2) 학습, 테스트 데이터 정확도가 계속 증가하는 경우(=과소적합) : 모델을 더 복잡하게 만들거나 학습횟수 늘려 학습을 충분히 시킬 필요가 있다.  <br>


3) 학습 데이터 정확도는 계속 증가하지만, 테스트 데이터 정확도는 증가하지 않는 경우(=과적합) : 모델을 단순화하거나 과적합 해소 기법 고려해보기. <br>



__오류 역전파__<br>
손실을 최소화하며 학습이 이뤄진다. 손실의 최소화는 미분을 통해 이루어지는데, 층이 깊어지면서 손실을 미분하는 계산은 복잡해지며 한번에 미분을 계산하기란 거의 불가능. 이를 해결하는 방법이 `오류 역전파`. 이는 함숫값은 정방향(forward)으로 계산하고 미분은 역방향(backward)으로 계산.

__규제강화__ <br>
과적합 개선을 위해 `규제강화` 방법 사용. <br>
-> 가중치에 제약을 주는 손실 추가, 데이터증강(augmentation), 모델을 단순화 하는 Dropout 등..

1. 손실추가 <br>
    - 손실 = 실제값 - 계산값
    - 가중치에 대한 규제(=행렬의 크기를 규제)를 추가해 계산. 

2. 데이터증강 <br>
    - keras의 경우 ImageDataGenerator 클래스를 사용해 데이터 증강할 수 있다. 

3. Dropout <br>
    - 신경망을 학습시키는 과정에서 과적합이 일어났다는 것은 학습모델이 데이터에 비해 복잡하다는 의미. 또한 신경망은 모든 뉴런들이 연결되어 있는 상태여서 복잡도가 높은 상태다. 그래서 신경망의 연결을 일정부분 제거함으로써 모델을 단순화해 과적합을 해결할 수 있다. 

__최적화기법__
1. 확률적 경사하강법 SGD의 한계
- SGD는 단순히 미분값을 구해 미분방향으로 학습률(learning_rate)만큼 가중치를 업데이트해주는 방식이다. 
- 하지만 노이즈가 많이 발생할 수 있고, 지그재그로 학습하여 값이 수렴하는데 계산을 여러번 수행해야 한다.
- 또한 극소값(local minimum)이나 안장점(saddle point)에서 미분값이 0이므로, 경사하강 진행하는 중 최솟값으로 진행하지 못하고 이 점들에 수렴할 수 있다. 더 이상 업데이트x

2. SGD의 개선
- 가중치를 업데이트할 때 **미분의 진행방향에 변화**를 주는 것.
    - `Momentum` : 미분방향에 다른 momentum을 추가해 고착되는 상황(local minimum문제)에 변화를 줌으로써 그 상황에서 벗어나게 함. 
    - `NAG` : 먼저 진행한 후 미분을 적용

- 가중치를 업데이트할 때 **이동 보폭(step size)에 변화**를 주는 것. 
    - 미분방향으로 일정한 속도(learning_rate)로 업데이트를 진행하다 보면 최솟값 근처에서 이동 보폭이 수렴하기에 너무 커져 노이즈가 심해지면서 수렴하는 데 오래 걸린다.
    - 따라서 처음에는 더 크게 움직여서 최소점 근처로 빨리 이동하고, 최소점 근처에 도달하면 점점 더 보폭이 줄어들어야 한다. 이 보폭 조절은 결국 학습이 진행되면서 학습률이 적당히 줄여간다는 의미
    - 이렇게 자동으로 보폭(learning rate)을 조절하는 방법을 사용한 것이 AdaGrad, RMSProp기법이다. 

- **진행방향변화 + 이동보폭 변화 = 최적화 함수는 Adam**

### 1. 패키지 설치 및 데이터 로드

```python
import numpy as np
import pandas as pd

from tensorflow.keras.datasets.mnist import load_data

(train_x, train_y), (test_x, test_y) = load_data()
```

```python
print(train_x.shape, train_y.shape)

print(test_x.shape, test_y.shape)
```

### 2. 데이터 확인

```python
from PIL import Image
img = train_x[0]

import matplotlib.pyplot as plt
# fromarray : numpy 배열로부터 PIL 이미지 객체 생성
# 즉, 주어진 Numpy배열 img를 사용해 8비트 그레이 스케일 이미지 생성(mode='L')
img1 = Image.fromarray(img, mode='L')
plt.imshow(img1)
train_y[0] # 0번째 학습용 데이터의 레이블 확인
```

### 3. 데이터 전처리

```python
# 데이터의 형태를 벡터로 변환
train_x1 = train_x.reshape(60000, -1)
test_x1 = test_x.reshape(10000, -1)
```

```python
train_x1[0]
```

```python
# 이미지 데이터이므로 255를 나눠준다.
train_x2 = train_x1 / 255
test_x2 = test_x1 / 255
```

```python
train_x2[0]
```

### 4. 모델 설정

```python
# 모델 설정에 필요한 라이브러리 호출
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# 모델명을 md로 설정해 선형층(Dense) 쌓기
md = Sequential()
md.add(Dense(128, activation='relu', input_shape=(28*28,)))
md.add(Dense(64, activation='relu'))
md.add(Dense(10, activation='softmax')) # 선형분류이므로 마지막 층은 softmax로 활성화
md.summary() #모델을 요약해 출력
```

> softmax는 다중클래스 분류 문제에서 출력 뉴런의 값들을 확률로 해석할 수 있게 해주며, 모든 출력 뉴런의 값의 합이 1이 되도록 한다. (= 모델이 각 클래스에 속할 확률을 출력)

### 5. 모델 학습 진행

```python
# 모델 compile
md.compile(loss='sparse_categorical_crossentropy', optimizer='sgd', metrics='acc')
```

> compile() : 모델을 학습하기 전에 모델을 구성.
> loss='sparse_categorical_crossentropy' : 정수 형태의 레이블을 다루며 다중 클래스 분류문제에 적합한 손실함수. 

```python
# 모델 학습 : 학습횟수, batch_size, validation_split설정
hist = md.fit(train_x2, train_y, epochs=50, batch_size=64, validation_split=0.2)
```

> batch_size=64 : 모델 학습 시 각 배치에 64개의 샘플을 사용. 즉, 64개의 샘플마다 가중치를 업데이트.
> validation_split=0.2 : 학습데이터의 20%를 검증 데이터로 사용. 나머지 20%는 각 에폭 후 모델의 성능을 검증하는데 사용. 

```python
hist.history
```

```python
# 학습 분석 그래프 그리기
acc = hist.history['acc']
val_acc = hist.history['val_acc']
epoch = np.arange(1, len(acc)+1)

plt.figure(figsize=(10,8))
# plt.xlim(250, len(acc)+1)
# plt.ylim(0.92, 0.94)

plt.plot(epoch, acc, 'b', label='acc')
plt.plot(epoch, val_acc, 'g', label='val_acc')
plt.legend()

plt.show()
```

> 과적합

### 6. 결과확인

```python
md.evaluate(test_x2, test_y)
```

> 생성된 모델 md와 실제값 test_x2, test_y와의 손실값과 정확도를 평가한다. <br>
> [0.09037388861179352, 0.9749000072479248] -> 첫번째 값은 손실값, 두번째 값은 정확도

#### 2) dropout 적용 모델

```python
# 모델 생성
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

md = Sequential()
md.add(Dense(128, activation='relu', input_shape=(28*28, )))
md.add(Dense(64, activation='relu'))
md.add(Dropout(0.5))
md.add(Dense(10, activation='softmax'))
md.summary()
```

> Dropout은 학습 중에 신경망의 일부 뉴런을 무작위로 선택하여 비활성화시키는 기법입니다. 간단히 말해, 학습 중에 네트워크의 일부를 임의로 끄는 것입니다. 이를 통해 네트워크가 특정 뉴런에만 의존하지 않고, 여러 가지 다양한 특징을 학습하도록 도와주며, 모델의 일반화 성능을 향상시킵니다. <br>
Dropout은 일반적으로 완전 연결(fully connected) 레이어 또는 컨볼루션(convolutional) 레이어의 뉴런에 적용됩니다. 학습 중에 각 뉴런을 무작위로 선택하여 제외하고, 제외된 뉴런은 해당 학습 배치(batch)에서 비활성화됩니다. 이것은 네트워크의 특정 부분이 다른 부분에 의존하지 않도록 하고, 각 뉴런이 독립적으로 유용한 특징을 학습하도록 돕습니다. <br>
Dropout은 모델이 학습할 때만 적용되며, 테스트나 평가 과정에서는 비활성화되어야 합니다. 이는 테스트 과정에서 네트워크가 모든 특징을 고려하고, 최종적으로 예측을 하도록 해야하기 때문입니다. <Br>
일반적으로 Dropout은 신경망의 성능을 향상시키고 과적합을 방지하는 데 매우 효과적인 기법 중 하나로 평가됩니다.

```python
# 모델 학습
md.compile(loss='sparse_categorical_crossentropy', optimizer='sgd', metrics='acc')
hist = md.fit(train_x2, train_y, epochs=50, validation_split=0.2, batch_size=64)
```

```python
hist.history
```

```python
acc = hist.history['acc']
val_acc = hist.history['val_acc']
epochs = np.arange(1, len(acc)+1)

plt.figure(figsize=(10,8))
plt.plot(epochs, acc, 'b', label='acc')
plt.plot(epochs, val_acc, 'g', label='val_acc')
plt.legend()

plt.show()
```

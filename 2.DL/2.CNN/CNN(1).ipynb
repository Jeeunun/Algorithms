# __CNN (Convolution Neural Networks)__
- 이미지 분류는 컴퓨터가 사람처럼 사물을 시각적으로 인식하고 분류해내는 작업을 의미한다. 이 분야를 Computer Vision이라 표현한다. 사람의 신경망을 형상화해서 만든 신경망 모델은 기본 단위가 선형모델이다. 신경망 모델보다 더 고도화된 분석기법이 바로 CNN !


__1. CNN모델은 기본 단위가 합성곱 필터이다.__ <br>
CNN모델은 층을 깊게 하면 이미지의 특성들이 고스란히 필터에 의해 걸러지게 된다. <br>
전이학습(Transfer Learning)은 적은 데이터로 학습을 진행할 때 **미리 학습된 모델의 가중치를 사용**해 학습에 활용하는 기법.

__2. CNN의 특징__ <br>
여러 층을 사용한 CNN 모델은 층이 깊어질수록 처리되는 과정이 구체적인 이미지에서 추상적인 것으로 변하는 것으로 관찰. 


__3. CNN의 구성요소__ <br>
 <br>

- Convolution Layer, Pooling Layer, FC(fully connected) Layer 로 구성되어 있다. <br>

-__1. Convolution Layer__ : 주어진 이미지(또는 텐서)를 **합성곱(convolution)으로 처리해 활성화 함수를 사용**해 활성화하는 과정 <br>

- (a) Convolution 연산
    - 합성곱은 텐서와 텐서 사이에서 정의되는 연산. <br>
    -  <br>
    -  <br>


- (b) Filter <br>
    - filter : 입력 이미지에 특정한 사이즈의 텐서(filter)를 사용해 전체 이미지의 일부를 스캔하듯이 이동하면서 연산. <br>
    - 보통 3 x 3 크기의 필터를 많이 사용. 필터의 두께는 자동으로 입력 데이터(텐서)의 두께로 설정. (만약 입력 이미지가 32*32*`3` 크기의 텐서라면 filter는 3*3*3으로 설정) <br>

- (c) feature map <br>
    - 입력 데이터에 필터로 스캔한 결과로 만들어지는 텐서를 feature map(특성맵)이라 한다. <br>
    - 보통 하나의 합성곱 층에서 여러 개의 필터가 사용되며, 그 결과로 필터 수만큼의 특성 맵이 만들어진다. <br>
    - 합성곱 연산(input data) -> 특성 맵 연산(feature map) -> 활성화함수 맵(activation map) <br>

- (d) Stride  <br>
    - 합성곱 연산을 할 때 필터가 움직이는 간격. <br>
    - 합성곱이 2차원(3x3)으로 정의되므로 stride도 2차원(m,n)으로 정의 : 한 번 합성곱 연산을 한 후에 우측으로 m만큼씩 이동해 입력 텐서의 끝까지 이동한 후, 아래로 n만큼씩 움직여서 맨 왼쪽부터 다시 스캔하는 방식. <br>
    -  <br>

- (e) padding  <br>
    - 입력 데이터에 합성곱 연산을 통해 만들어진 특성맵은 원래 입력 데이터의 크기보다 작아지게 된다. 만약 입력 이미지의 크기가 작은 경우라면 몇 번의 합성곱을 실행하면 데이터의 크기가 작아져 더 이상 합성곱 층을 추가하는 것이 불가능하다. 그래서 **합성곱 계산을 할 때 padding을 덧대어 출력 크기를 입력 데이터의 크기와 같게 만들 수 있다.** 이때 합성곱 층에서 인수설정을 padding=same으로 사용하면 된다.
    -  

- (f) 활성화 (Activation) <br>
    - 활성화 함수는 주로 ReLU를 사용. <br>
    - 활성화 함수의 역할 (=신경망의 비선형성 도입) :
        - 활성화 함수 없이 CNN을 구성하면, 각 층(layer)은 사실상 선형변환(linear transformation)을 수행하는 것에 불과하다. 즉, 여러층을 겹쳐도 결국은 하나의 선형 변환으로 단순화될 수 있다. **비선형 활성화 함수를 사용함으로써 단순화되는 문제를 해결하고 신경망이 복잡한 데이터의 특징과 패턴을 학습할 수 있게 된다.**

-__2. Pooling Layer__ : 합성곱으로 처리된 결과를 **정해진 단위별로 모아 특성을 강화**시키는 과정. <br>
- (a) MaxPooling2D <br>
    - 입력 데이터에서 정해진 크기(보통(2,2))에서 최댓값을 출력하는 Pooling. <br>
    -  <br>

- (b) AveragePooling2D <br>
    -  <Br>

- (c) GlobalAveragePooling2D <br>
    - 합성곱 연산으로 만들어진 하나의 특성 맵에서 평균값을 출력. 데이터의 크기를 더 많이 줄이게 된다. <Br>
    - GoogLeNet에서 Fully Connected Layer 직전에 Flatten 대신 사용되었다. <br>
    -  <br>


-__3. FC Layer__ : 이전 단계의 결과 텐서를 **벡터화(Flatten)**해 최종적으로 출력하고자 하는 결과 사이를 신경망으로 연결시켜주는 역할. 마지막에 출력값을 조절해주는 역할  <br>


여기서 CNN모델에만 있는 합성곱 층과 Pooling Layer로 이루어진 층을 CNN Layer하 한다. 이 CNN Layer를 여러 번 반복 사용해 모델을 설계하며, 끝부분에는 항상 FC Layer로 출력 층을 설정. 또한 CNN Layer는 합성곱 층을 보통 1~3번 사용한 후 Pooling Layer를 한 번 적용한다. CNN Layer와 FC Layer 사이에는 텐서를 벡터로 만들어주는 벡터화 과정이 필요하다. 이 과정에는 주로 Flatten 클래스를 사용. 

__4. CNN 과정__ <br>


__Step1. 이미지 데이터 불러오기__ <br>
- 방법1: 이미지 데이터를 numpy 등의 숫자 데이터로 변환해 전체 데이터를 저장해서 학습시키는 것. <br>
    - 이미지 데이터의 용량이 큰 경우는 첫번째 방법으로 이미지를 변환해 저장하기에는 너무 큰 용량이 필요하므로 저장이 불가능하거나, 저장하더라도 효율적인 학습이 이루어지지 않게 된다. <br>
- 방법2: 이미지 자체를 적당한 개수(batch_size)만 불러와 사용. 
    - keras.ImageDataGenerator 사용

- __ImageDataGenerator 클래스__
기존의 이미지를 그대로 불러올 수 있고, 크기나 색상을 변환하거나 전처리, 데이터증강에 사용하는 크기조절, 확대, 축소, 회전 등 할 수도 있다. __단순히 이미지를 불러오는 역할만 할 때는 데이터 전처리만 적용해 rescale=1/255 인수만 추가하면 된다.__

- 1. flow_from_directory
    - 디렉토리 구조에서 데이터를 불러오는 함수. 이 함수는 **이미지 데이터가 이미 class 별로 디렉토리 형태로 정리되어 있는 경우에 사용이 가능**. 레이블은 디렉토리 정보를 활용해 자동으로 one-hoe encoding 된다. 

 - 2. flow_from_dataframe
    - 이 함수는 **이미지 파일 경로와 이미지 레이블이 DataFrame 파일에 정리되어 있는 경우에 사용이 가능.** DataFrame파일에서 이미지 파일의 경로와 레이블을 불러와 batch_size별로 데이터를 출력한다.

__6. 전이학습 Transfer Learning__ <br>
이미 학습된 모델을 가져와서 출력 부분만 변경해 새로운 데이터를 학습시키는 방법 <br>
기존에 성능이 높은 모델을 처음부터 학습시키려면 많은 데이터와 학습시간이 필요한데, **전이학습을 사용하면 적은 데이터를 사용하고도 좋은 성능을 발휘할 수 있으며 학습 시간도 많이 단축**시킬 수 있다. 

-1) 특성추출 <br>
    - 기존에 학습된 모델의 가중치를 그대로 가져오는 과정.기존 학습된 모델의 가중치를 불러와 사용할 부분을 그대로 고정하고, 학습시킬 부분은 주어진 데이터에 맞게 분리.

-2) 미세조정 <br>
    - 원래 모델에서 특성 추출 시 고정한 부분 외에 새로 학습시킬 부분의 가중치와 새로운 데이터 및 목적에 맞게 FC Layer를 다시 설계해 학습시키는 과정. <br>
    - 학습의 큰 틀은 이미 특성 추출 시 고정했기에 적은 수의 가중치만을 학습시킴으로써 충분히 좋은 결과를 얻을 수 있다. 

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(rescale=1./255)
```

```python
ImageDataGenerator(rotation_range=0, width_shift_range=0.0, height_shift_range=0.0, 
                   brightness_range=None, shear_range=0.0, zoom_range=0.0, 
                   horizontal_flip=False, vertical_flip=False,)
```

> 해당 인수들을 적절히 사용하면 충분히 많은 양의 데이터를 증강해 사용할 수 있다. <br>

height_shift_range : 부동 소수점, 1D형태의 유사배열 혹은 정수<br>
brightness_range :  두 부동 소수점 값으로 이루어진 리스트 혹은 튜플, 밝기 정도를 조절할 값의 범위<br>
shear_range : 부동소수점. 층 밀리기의 강도 <br>
zoom_range : 부동소수점 혹은 하한.상산. 무작위 줌의 범위<br>
horizontal_flip : 불리언. 인풋을 무작위로 가로로 뒤집음<br>
vertical_flip : 불리언. 인풋을 무작위로 세로로 뒤집음<br>
